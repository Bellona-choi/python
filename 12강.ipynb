```python
#파일 다루기
f=0 
try:
    f=open("LICENSE.txt","rt") #읽기,텍스트 모드로 파일 열기
    while True:
        r=f.readline() #연 파일을 한줄씩 읽기
        if not r:#만약 더 이상 읽을 줄이 없다면
            break #반복문 끝내기
        print(r)      
except FileNotFoundError: #FileNotFoundError에 대해서만 except 구문을 실행 
    print("파일없음")
finally:
    if(f): f.close() #만약 파일을 닫는다면
    print("프로그램 끝")
```

```python
#파일 쓰기
import random as r
f=0
f=open("score.txt","wt") #파일 쓰기모드 및 텍스트모드로
for i in range(100):#100까지 
    one_person="홍길동"+str(i)+","+str(r.randrange(0,100))+","+str(r.randrange(0,100))+","+str(r.randrange(0,100))+"\n" 
    #홍길동(번호)+랜덤 점수 + 랜덤점수 + 렌담점수 출력
    f.write(one_person) #한 줄씩 쓴다 
    print(one_person)
f.close()#파일 닫기
```

```python
#파일 읽기
f=0
f=open("score.txt","rt") 
while True:
    one_person=f.readline() #한 줄씩 읽는다
    if not one_person: break #읽을 내용이 없으면 루프탈출
    one_person=one_person.replace("\n","") #줄 바꿈문자 없애기 
    item=one_person.split(",") #콤마 구분자로 짜르기
    print("="*10)
    print("이름: ", item[0])
    print("국어: ", item[1])
    print("영어: ", item[2])
    print("수학: ", item[3])
    print("="*10) 
f.close()
```

```python
#파이썬에서sqlite3 사용 실습
import sqlite3 as s
con=s.connect("my01.db")#my01.db와 연결
cursor = con.cursor()#con의cursor함수를 cursor로 저장
con.commit()#con 실행
#table 생성
cursor.execute("create table examtable(name varchar(20),studentid int not null primary key,kor int,eng int,mat int)")
con.commit()#버퍼에 처리된 내용을 기록 
cursor.close()#cursor 닫기
con.close()#con 닫기
```

```python
con=s.connect("my01.db") 
cursor = con.cursor()#con의cursor함수를 cursor로 저장
cursor.execute("select * from examtable")#테이블 내용 출력
one_rec=cursor.fetchall()  #해당 데이터를 전부 가져옴
for field in one_rec:  #해당 데이터를 반복문을 통해 하나씩 처리 
    kor=field[2]#국어 호출
    eng=field[3]#영어 호출
    mat=field[4]#수학 호출
    print("이름:%s 학번:%s 국어:%3d 영어:%3d 수학:%3d 총점:%3d 평균:%3d"%(field[0],field[1],kor,eng,mat,kor+eng+mat,int((kor+eng+mat)/3))) 
cursor.close()#cursor 닫기
con.close()#con 닫기
```
